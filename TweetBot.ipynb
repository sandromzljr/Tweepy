```python
import tweepy
```

```python
#Cred√™nciais necess√°rias para consumir api do Twitter
bearerToken = ''
consumerKey = ''
consumerSecret = ''
accessToken = ''
accessTokenSecret = ''
```

```python
#Autentica√ß√£o 
twitter = tweepy.Client(bearer_token=bearerToken, consumer_key=consumerKey, consumer_secret=consumerSecret, access_token=accessToken, access_token_secret=accessTokenSecret)
```

```python
#Buscando tweets
resultado = twitter.search_recent_tweets(query='monark', max_results=100)
tweets = resultado.data
```

```python
#Retornando os tweets e demarcando o fim deles.
for tweet in tweets:
    print(tweet)
    print('-----')
```
